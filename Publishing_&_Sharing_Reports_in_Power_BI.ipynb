1. Explain why publishing and sharing reports is considered more important than just building dashboards
in Power BI Desktop

- Building dashboards in Power BI Desktop is only the first step. The real value comes from sharing insights.

Dashboards created locally are not accessible to others.
Publishing allows collaboration and decision-making across teams.
Stakeholders (managers, analysts) need real-time access.
Sharing ensures data-driven culture in organizations.

 Without publishing, reports remain personal; with sharing, they become organizational assets.


2.  Define Power BI Service and explain its role in comparison to Power BI Desktop in the reporting lifecycle.
- Power Bi Service is a cloud-based platform used to publish, share, and collaborate on reports
Comparison with Power BI Desktop

| Feature      | Power BI Desktop | Power BI Service    |
| ------------ | ---------------- | ------------------- |
| Type         | Development tool | Cloud platform      |
| Purpose      | Create reports   | Share & collaborate |
| Usage        | Individual       | Organization-wide   |
| Data refresh | Manual           | Scheduled/automatic |

Role in lifecycle:

Desktop → Build
Service → Publish, Share, Secure, Collaborate

9. Explain Row-Level Security (RLS) in Power BI.
How does RLS ensure that different users see different data using the same report?
- RLS restricts data access based on user roles.

How it works:
Define roles (e.g., Region = “North”)
Assign users to roles
When user opens report → sees only their data

Example:

Manager A → sees North region
Manager B → sees South region

Same report, different data visibility.
